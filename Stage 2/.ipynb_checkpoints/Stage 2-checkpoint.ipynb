### Doing a swarm plot
https://www.kaggle.com/ndrewgele/visualizing-pok-mon-stats-with-seaborn

```python
pkmn = pd.melt(data2, id_vars=["Name", "Type 1", "Type 2"], var_name="Stat")
pkmn
```

```python
sns.swarmplot(x="Stat", y="value", data=pkmn, hue="Type 1");
```

Improved version of the swarmplot

```python
sns.set_style("whitegrid")
with sns.color_palette([
    "#8ED752", "#F95643", "#53AFFE", "#C3D221", "#BBBDAF",
    "#AD5CA2", "#F8E64E", "#F0CA42", "#F9AEFE", "#A35449",
    "#FB61B4", "#CDBD72", "#7673DA", "#66EBFF", "#8B76FF",
    "#8E6856", "#C3C1D7", "#75A4F9"], n_colors=18, desat=.9):
    plt.figure(figsize=(12,10))
    plt.ylim(0, 275)
    sns.swarmplot(x="Stat", y="value", data=pkmn, hue="Type 1", split=True, size=7)
    plt.legend(bbox_to_anchor=(1, 1), loc=2, borderaxespad=0.);
```

```python
ax = sns.violinplot(x="Stat", y="value", data=pkmn, inner=None)
ax = sns.swarmplot(x="Stat", y="value", data=pkmn, color="white", edgecolor="black")
```

```python
#Extract unique types of Pokemon
typesPkm = pkmn['Type 1'].unique()
typesPkm
```

```python
grass = pkmn['Type 1'] == 'Grass'
pkmn[grass]
```

```python
fig, axs = plt.subplots(ncols=18)
count = 0 
for x in typesPkm:
    typePok = pkmn['Type 1'] == x
    dataPok = pkmn[typePok]
    ax = sns.violinplot(x="Stat", y="value", data=dataPok, inner=None, ax=axs[count])
    ax = sns.swarmplot(x="Stat", y="value", data=dataPok, color="white", edgecolor="black", ax=axs[count])
    count = count+1
```

```python
import numpy as np
types = pkmn['Type 1']

colors = [
    'yellowgreen',
    'gold',
    'lightskyblue',
    'lightcoral',
    'yellow',
    'lightgreen',
    'silver',
    'white',
    'pink'
]


explode = np.arange(len(types.unique())) * 0.01
plt.figure(figsize=(20,30))
types.value_counts().plot.pie(
    explode=explode,
    colors=colors,
    title="Types of Pokemon",
    autopct='%1.1f%%',
    shadow=True,
    startangle=90,
    figsize=(9,8)
)
plt.title("Percentage Types of Pokemon", y=1.1)
plt.tight_layout()
```
